```python
grades = pd.read_csv("grades.csv")
```

```python
grades.head()
```

```python
grades["submit_time"] = pd.to_datetime(grades["submit_time"])
grades.head()
```

```python
grades["format"] = grades["submission"].apply(lambda x: x.split(".")[-1])
grades.head()
```

```python
100 * (grades.format == "zip").sum() / grades.shape[0]
```

```python
grades.format.value_counts()
```

```python
100 * 189 / 198
```

```python
# deadline Jan 3, 2017 - 11:59:59 
## 2017-01-03 23:59:59

(grades["submit_time"] > "2017-01-03 23:59:59").sum()
```

```python
grades[grades["submit_time"] > "2017-01-03 23:59:59"].shape
```

```python
grades["date"] = grades["submit_time"].dt.date
```

```python
grades.groupby(by="date")["submit_time"].count()
```

```python
grades["date"].value_counts()
```

```python
grades["submit_time"].dt.hour.value_counts()
```

```python
grades["hour"] = grades["submit_time"].dt.hour
```
